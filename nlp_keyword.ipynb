```python
def transcript_capture(video_id, filename):
    transcript = YouTubeTranscriptApi.get_transcript(video_id)

    formatter = WebVTTFormatter() #change this line of code to JSONFormatter() if want it in json file

    vtt_formatted = formatter.format_transcript(transcript)

    with open(filename, 'w', encoding = 'utf-8') as vtt_file:
        vtt_file.write(vtt_formatted)
```

```python
def tokenize_col(df, col):
    # new_col = [nltk.word_tokenize(df[col][0])]
    #new_col = ["a"]
    new_col = []
    for i in range(len(df[col])):
        new_col.append(nltk.word_tokenize(df[col][i]))
    return new_col
```

```python
def whole_transcript(df, col):
    new_col = []
    for i in range(len(df[col])):
        new_col.extend(df[col][i])
    return new_col
```

```python
paul_rogers_whole_transcript = whole_transcript(paul_rogers_short, "Tokens")
```

```python
stop_words=set(stopwords.words("english"))
def filtering_stopwords(transcript):
    filtered_transcript = []
    for w in transcript:
        if w not in stop_words:
            filtered_transcript.append(w)
    return filtered_transcript
```

```python
def hashtags(components, feature_names, vid_name):
  with open('/content/drive/MyDrive/The Berkeley Forum/6. Committees and Divisions/Systems and Technology Committee/SP 24/nlp keyword extraction/hashtags/'+vid_name+'.txt', 'w') as writefile:
    for topic in components:
      for i in topic.argsort()[:-num_keywords - 1:-1]:
        writefile.write("#" + feature_names[i] + " ")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF


# Load the dataset
# data = pd.read_csv(sanjay.df)
paul_rogers_short_event = filtering_stopwords(paul_rogers_whole_transcript)

# Vectorize the text using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english', max_df=3, min_df=0)
tfidf = vectorizer.fit_transform(paul_rogers_short_event)


# Apply Non-Negative Matrix Factorization (NMF) for topic modeling
nmf_model = NMF(n_components=5, random_state=42)
nmf_model.fit(tfidf)


# Get the top 10 keywords for each topic
num_keywords = 10
feature_names = vectorizer.get_feature_names_out()
for topic_idx, topic in enumerate(nmf_model.components_):
    print("Topic %d:" % (topic_idx))
    print(", ".join([feature_names[i] for i in topic.argsort()[:-num_keywords - 1:-1]]))
    print("\n")
hashtags(nmf_model.components_, feature_names, "paul_rogers_3_min")
```
